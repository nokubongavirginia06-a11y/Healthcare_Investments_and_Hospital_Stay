## Univariate analysis

### Subtask:
Visualize the distribution of individual features using histograms, box plots, and count plots.


**Reasoning**:
Generate histograms and box plots for numerical columns and a count plot for the categorical column to visualize the distribution of individual features.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Time', 'Hospital_Stay', 'MRI_Units', 'CT_Scanners', 'Hospital_Beds']
categorical_col = 'Location'

# Histograms for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data=df, y=col)
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()

# Count plot for categorical column
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x=categorical_col)
plt.title(f'Count of {categorical_col}')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Bivariate analysis

### Subtask:
Explore relationships between pairs of features using scatter plots, heatmaps, or other appropriate visualizations.


**Reasoning**:
Create scatter plots for 'Hospital_Stay' against the investment-related columns, calculate the correlation matrix, and visualize it using a heatmap.



```python
# Scatter plots
numerical_investment_cols = ['MRI_Units', 'CT_Scanners', 'Hospital_Beds']

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_investment_cols):
    plt.subplot(1, 3, i + 1)
    sns.scatterplot(data=df, x=col, y='Hospital_Stay')
    plt.title(f'Hospital Stay vs {col}')
plt.tight_layout()
plt.show()

# Correlation matrix
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix for numerical features.


**Reasoning**:
Calculate the correlation matrix for the numerical features and visualize it using a heatmap with annotations and a title.



```python
numerical_cols = ['Time', 'Hospital_Stay', 'MRI_Units', 'CT_Scanners', 'Hospital_Beds']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
